```python
import pandas as pd
import requests
import os
#combine 2019-20 and 2020-21 Team Understat Data
s19_20=pd.read_csv("2019-2020_Team_Understat.csv")
s20_21=pd.read_csv("2020-2021_Team_Understat.csv")
Combine=s20_21.append(s19_20)
Combine
#os.getcwd() - checks my current working directory
#os.listdir(os.getcwd()) - tells me what is in my directory to use relative path
```

```python
#Aggregate data into Averages

#return average score,xG,npxG,conceded,xGA,npxGA by team
Agg1=Combine.groupby('team')['scored','xG','npxG','missed','xGA','npxGA']
t1=Agg1.mean()

#return number of games by team
Agg2=Combine.groupby('team')['date']
t2=Agg2.count()


Agg3=Combine.groupby('team')['pts','xpts']
t3=Agg3.sum()

frame=[t2,t1,t3]
Agg4=pd.concat(frame,axis=1)
Agg4.rename(columns = {'date':'played'}, inplace = True)
Agg4
```

```python
#Estimate Brentford's xG,npxG,xGA,npxGA using Southampton + Last 3 Relegated Teams (Sheffield United, Fulham, West Brom)

Estimates=['Southampton','Sheffield_United','West_Bromwich_Albion','Fulham']
Season=['20-21']
Base=Combine[Combine.team.isin(Estimates) & Combine.Season.isin(Season)]
Base['team']='Brentford'

#return average score,xG,npxG,conceded,xGA,npxGA for Brentford
Agg1=Base.groupby('team')['scored','xG','npxG','missed','xGA','npxGA']
t1=Agg1.mean()

#return number of games for Brentford
Agg2=Base.groupby('team')['date']
t2=Agg2.count()

#return total points for Brentford
Agg3=Base.groupby('team')['pts','xpts']
t3=Agg3.sum()

frame=[t2,t1,t3]
Agg5=pd.concat(frame,axis=1)
Agg5.rename(columns = {'date':'played'}, inplace = True)
Agg5['played']=Agg5['played']/4
Agg5['pts']=Agg5['pts']/4
Agg5['xpts']=Agg5['xpts']/4
Agg5
```

```python
Final_Agg=Agg4.append(Agg5)
Final_Agg_2=Final_Agg.reset_index()
Final_Agg_2
```

```python
#remove relegated teams
drop=['Fulham','West_Bromwich_Albion','Sheffield_United','Bournemouth']
Final=Final_Agg_2[~Final_Agg_2.team.isin(drop)].sort_values('team')
Final['id']=range(1,21)
Final
```

```python
Final.to_csv("2019-2021_Agg_Understat.csv")
```
