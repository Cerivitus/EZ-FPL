```python
import requests
import pandas as pd
import os
import numpy as np
#pulling me FPL 2021-2022 Fixture Data
response=requests.get("https://fantasy.premierleague.com/api/fixtures/")
j=response.json()
fixtures = pd.DataFrame.from_dict(j)
fixtures[['event','team_a','team_h']].sort_values('event')
save=pd.melt(fixtures, id_vars =['event','id'], value_vars =['team_a','team_h']).sort_values('id')
save
#convert variable to home or not
#bring teams
#bring opponents
#keep the event (matchday)


#reading teamID masterlist
#os.getcwd() - checks my current working directory
#os.listdir(os.getcwd()) - tells me what is in my directory to use relative path
```

```python
fixtures[['event','id','team_a']].sort_values('event')
```

```python
teamID=pd.read_csv('https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/master/data/2021-22/teams.csv')
cleanteamID=teamID[['id','name','short_name']]
cleanteamID
```

```python
save_gw=pd.merge(save,cleanteamID,left_on="value",right_on="id").sort_values('id_x')
save_gw2=save_gw[['event','id_x','variable','value','name','short_name']]
save_gw2.columns=['GW','id','home?','team_id','name','short_name']
team_h=save_gw2.loc[save_gw['variable'] == 'team_h']
team_a=pd.merge(team_h,fixtures[['event','id','team_a']],left_on=['GW','id'],right_on=['event','id'])
team_a=team_a.drop(columns=['event'])
team_a.columns=['GW','match_id','home?','team_id','t1','s1','oppo_id']
team_h=pd.merge(team_a,cleanteamID,left_on="oppo_id",right_on="id").sort_values('GW')
team_1=team_h.drop(columns=['id'])
team_1.columns=['GW','match_id','home?','team_id','t1','s1','oppo_id','t2','s2']
team_1



```

```python
team_h=save_gw2.loc[save_gw['variable'] == 'team_a']
team_a=pd.merge(team_h,fixtures[['event','id','team_h']],left_on=['GW','id'],right_on=['event','id'])
team_a=team_a.drop(columns=['event'])
team_a.columns=['GW','match_id','home?','team_id','t1','s1','oppo_id']
team_a=pd.merge(team_a,cleanteamID,left_on="oppo_id",right_on="id").sort_values('GW')
team_2=team_a.drop(columns=['id'])
team_2.columns=['GW','match_id','home?','team_id','t1','s1','oppo_id','t2','s2']
team_2
```

```python
final=pd.concat([team_1,team_2])
final.sort_values('match_id')
```

```python
#creating a boolean home team column in final
final['home_team?'] = np.where(final['home?']== 'team_h', True,False)
final_2=final.drop(columns=['home?']).sort_values('match_id')
final_2
```

```python
#bring SPI ranking data here and join to fixtures_names
#read in 538 SPI data (needs to scheduled after BPL matchday is completed)
teams = pd.read_csv("https://projects.fivethirtyeight.com/soccer-api/club/spi_global_rankings.csv")
#filtering on Barclays Premier League teams only
select_league='Barclays Premier League'
drop=['Fulham','West Bromwich Albion','Sheffield United']
add=['Brentford','Watford','Norwich City']

Base_BPL=teams[(teams.league==select_league)& ~teams.name.isin(drop)]
Relegated=teams[teams.name.isin(drop)]
Promoted=teams[teams.name.isin(add)]
Promoted["league"].replace({"English League Championship": "Barclays Premier League"}, inplace=True)
#add new promoted teams to EPL list
New_BPL=Base_BPL.append(Promoted)

#sort from A-Z
A_Z=New_BPL.sort_values(['name'])
A_Z.reset_index(inplace=True)
A_Z["id"] = A_Z.index + 1
#A_Z.to_csv('538-Data.csv')
# Select the columns I want
A_Z = A_Z[['id','rank','prev_rank','off','def','spi']]
A_Z
```

```python
# join final_2 and A_Z
Fixtures_538=final_2.merge(A_Z[['id','off','def','spi']],left_on='team_id',right_on='id')
Fixtures_538_1=Fixtures_538.drop(columns=['id'])
Fixtures_538_1.columns=['GW','match_id','team_id','t1','s1','oppo_id','t2','s2','home?','h_off','h_def','h_spi']
Fixtures_538_2=Fixtures_538_1.merge(A_Z[['id','off','def','spi']],left_on='oppo_id',right_on='id')
Fixtures_538_3=Fixtures_538_2.drop(columns=['id'])
Fixtures_538_3.columns=['GW','match_id','team_id','t1','s1','oppo_id','t2','s2','home?','team_off','team_def','team_spi','oppo_off','oppo_def','oppo_spi']

Fixtures_538_3.sort_values('match_id')
```

```python
Fixtures_538_3.to_csv('2021-2022_EPL_Fixtures.csv')
```
